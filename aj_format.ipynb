```python
import pandas as pd

# 读取CSV文件
df = pd.read_csv('load_P_allbus.csv')

# 获取第一列数据
column_data = df['allbus_p']

# 拆分数据成每16行一组
grouped_data = [column_data[i:i+16].tolist() for i in range(0, len(column_data), 16)]

# 创建新的DataFrame来存储拆分后的数据
new_df = pd.DataFrame(grouped_data, columns=[i+1 for i in range(16)])

# 打印结果
print(new_df)
```

```python
# 将结果保存到新的CSV文件
new_df.to_csv('allbus_p.csv', index=False)
```

```python
import pandas as pd

# 读取CSV文件
df = pd.read_csv('allbus_p.csv')

# 按照"Cluster"列的值进行升序排序
df_sorted = df.sort_values('Cluster')

# 打印排序后的结果
print(df_sorted)
# 将结果保存到新的CSV文件
df_sorted.to_csv('allbus_p_C1-C5.csv', index=False)
# 筛选"Cluster"列中数值为1的数据
filtered_data = df[df['Cluster'] == 1]
print(filtered_data)
```
